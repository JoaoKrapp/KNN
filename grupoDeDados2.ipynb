```python
from knn import KNN, load_data
import matplotlib.pyplot as plt
import pandas as pd
```

# Q2.1: Aplique seu kNN a este problema. Qual é a sua acurácia de classificação?

### Resposta: Acuracia maxima de 78,33% quando k=7

```python

q1data = []

def q1():
    train_features, train_labels, test_features, test_labels = load_data('./dados/grupoDados2.mat')
    
    knn = KNN(train_features, train_labels, test_features, test_labels)
    k_tests = knn.run_range_k(50)
    
    q1data.append(k_tests['accuracy'])
    
    plt.plot(k_tests['k'], k_tests['accuracy'])
    plt.show()
    
q1()
```

```python
for i in range(len(q1data[0])):
    print(f"{i} : {q1data[0][i]}")
```

# Q2.2: A acurácia pode ser igual a 98% com o kNN. Descubra por que o resultado atual é muito menor. Ajuste o conjunto de dados ou k de tal forma que a acurácia se torne 98% e explique o que você fez e por quê.

### Resposta: Utilizando somente as colunas 6, 0, 9, 4 conseguimos uma acuracia de 98,3334%, consegui esse resultado assim:
- Fiz o KNN para cada coluna separada
- Vi qual coluna tinha a maior acuracia( coluna 6)
- Fiz o KNN para cada coluna separada, mas adicionei o 6
- Vi que a a coluna que tinha a maior acuracia era o 0

Fui adicionando as colunas com maior acuracia até chegar no resultado

```python
q2data = {
	'column_removed' : [],
	'k' : [],
	'accuracy' : []
}

def q2():
	
	train_features, train_labels, test_features, test_labels = load_data('./dados/grupoDados2.mat')
	
	knn = KNN(train_features, train_labels, test_features, test_labels, columns_only=[4, 6, 0, 9])
	k_tests = knn.run_range_k(15)
	
	plt.plot(k_tests['k'], k_tests['accuracy'])
	plt.show()
		
		
q2()
```

```python
df_q2data = pd.DataFrame(q2data)
```

```python
teste = df_q2data.sort_values(by='accuracy', ascending=False)
```
