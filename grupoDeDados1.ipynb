```python
from knn import KNN, load_data
import matplotlib.pyplot as plt
```

# Q1.1. Qual é a acurácia máxima que você consegue da classificação?

### Resposta: 98%

```python
def q1():
    train_features, train_labels, test_features, test_labels = load_data('./dados/grupoDados1.mat')
    
    knn = KNN(train_features, train_labels, test_features, test_labels)
    k_tests = knn.run_range_k(15)
    
    plt.plot(k_tests['k'], k_tests['accuracy'])
    plt.show()
    
q1()
```

```python
q2_data = {}

def q2():
	train_features, train_labels, test_features, test_labels = load_data('./dados/grupoDados1.mat')
 
	for col in train_features.columns:
		
		knn = KNN(train_features, train_labels, test_features, test_labels, columns_to_remove=[col])
	
		k_tests = knn.run_range_k(15)
		q2_data[col] = k_tests
  
def q2plot():
	plt.figure(figsize=(20, 4))

	for coluna_removida in q2_data.keys():
		ax = plt.subplot(1, 4, coluna_removida + 1)
		ax.plot(q2_data[coluna_removida]['k'], q2_data[coluna_removida]['accuracy'])
	
		ax.set_xlabel("Neightbors")
	
	plt.show()

	
```

```python
q2()
```
